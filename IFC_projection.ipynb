```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")

# Input Data
data = {
    'Portfolio_Outstanding': [
        413310637, 4207813455, 7433078280, 10058329867, 10687684207, 10912470344
    ],
    'Delinquency_Over_360': [
        57060935, 144170057, 110762746, 478585832, 1147947984, 1518757635
    ],
    'Number_of_Loans': [409, 3024, 5220, 7759, 9128, 9151]
}

years = [2020, 2021, 2022, 2023, 2024, 2025]
df = pd.DataFrame(data, index=years)
df.index.name = 'Year'
df = df.reset_index()  # Make 'Year' a column

# Features and targets
X = df[['Year']]
y_portfolio = df['Portfolio_Outstanding']
y_delinq = df['Delinquency_Over_360']
y_loans = df['Number_of_Loans']

# Scale features (important for SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "SVM (SVR)": SVR(kernel='rbf', C=100, gamma='scale'),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42)
}

# Future years
future_years = np.array([2026, 2027, 2028, 2029]).reshape(-1, 1)
future_years_scaled = scaler.transform(future_years)  # Scale future years for SVM

# Store results
results = []
forecasts = {"Year": future_years.flatten()}

# Evaluate each model
for name, model in models.items():
    # Use scaled or unscaled X based on model
    X_train = X_scaled if name == "SVM (SVR)" else X.values

    # Cross-validation score (R²)
    try:
        cv_scores_portfolio = cross_val_score(model, X_train, y_portfolio, cv=3, scoring='r2')
        r2_portfolio = cv_scores_portfolio.mean()
    except:
        r2_portfolio = np.nan

    try:
        cv_scores_delinq = cross_val_score(model, X_train, y_delinq, cv=3, scoring='r2')
        r2_delinq = cv_scores_delinq.mean()
    except:
        r2_delinq = np.nan

    try:
        cv_scores_loans = cross_val_score(model, X_train, y_loans, cv=3, scoring='r2')
        r2_loans = cv_scores_loans.mean()
    except:
        r2_loans = np.nan

    # Train and forecast
    model.fit(X_train, y_portfolio)
    X_pred = future_years_scaled if name == "SVM (SVR)" else future_years
    portfolio_pred = model.predict(X_pred)

    model.fit(X_train, y_delinq)
    delinq_pred = model.predict(X_pred)

    model.fit(X_train, y_loans)
    loans_pred = model.predict(X_pred)

    # Store forecasts
    forecasts[f"{name}_Portfolio"] = np.round(portfolio_pred).astype(int)
    forecasts[f"{name}_Delinquency"] = np.round(delinq_pred).astype(int)
    forecasts[f"{name}_Loans"] = np.round(loans_pred).astype(int)

    # Save R² scores
    results.append({
        "Model": name,
        "Portfolio R²": f"{r2_portfolio:.3f}" if not np.isnan(r2_portfolio) else "N/A",
        "Delinquency R²": f"{r2_delinq:.3f}" if not np.isnan(r2_delinq) else "N/A",
        "Loans R²": f"{r2_loans:.3f}" if not np.isnan(r2_loans) else "N/A"
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Convert forecasts to DataFrame
forecast_df = pd.DataFrame(forecasts)
```

```python
print("🏆 MODEL ACCURACY COMPARISON (Cross-Validated R² Score)")
print("="*70)
print(results_df.to_string(index=False))
```
